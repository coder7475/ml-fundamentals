# Web Scaping Using Beautiful Soup

```python
# Scraping from bbc bangla
from urllib.request import Request, urlopen

myurl = "https://www.bbc.com/bengali"

# Set a user-agent header to avoid HTTP 403 Forbidden error
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
req = Request(myurl, headers=headers)

# Query the website so that it returns html page
html = urlopen(req).read()
html
```

```python
# Open the scraped page with beautiful soup
from bs4 import BeautifulSoup

soupified = BeautifulSoup(html, 'html.parser')
soupified
```

```python
# Print 2000 chars to get an idea about the page
from pprint import pprint

pprint(soupified.prettify()[:2000])
```

```python
# Get the title of the webpage
soupified.title.contents[0]
```
